```python
# Simple Rule-Based Chatbot

import re
import random
from datetime import datetime

print("🤖 Hello! I am RuleBot.")
print("Type 'exit' to end the chat.\n")

# Predefined responses
greetings = ["hello", "hi", "hey", "good morning", "good evening"]
bye_words = ["bye", "exit", "quit", "goodbye"]

def get_time():
    """Returns current system time"""
    return datetime.now().strftime("%I:%M %p")

def match_pattern(user_input, patterns):
    """
    Checks if any keyword in patterns exists in user input
    Returns True if matched
    """
    for word in patterns:
        if re.search(r"\b" + word + r"\b", user_input):
            return True
    return False


while True:

    user_input = input("You: ").lower().strip()

    # Exit condition
    if match_pattern(user_input, bye_words):
        print("RuleBot: Goodbye! Have a great day 😊")
        break

    # Greeting
    elif match_pattern(user_input, greetings):
        print("RuleBot:", random.choice([
            "Hello there! 👋",
            "Hi! How can I help you today?",
            "Hey! Nice to talk to you 😊"
        ]))

    # Ask name
    elif "your name" in user_input:
        print("RuleBot: I am RuleBot, your simple rule-based assistant.")

    # Ask time
    elif "time" in user_input:
        print("RuleBot: The current time is", get_time())

    # Ask about Python
    elif "python" in user_input:
        print("RuleBot: Python is a beginner-friendly programming language used for AI, data science, and web development.")

    # Ask about help
    elif "help" in user_input:
        print("RuleBot: I can respond to greetings, tell time, talk about Python, and more!")

    # Default response
    else:
        print("RuleBot:", random.choice([
            "Hmm... I am not sure about that.",
            "Can you rephrase your question?",
            "I am still learning new rules 😊"
        ]))
```
